# Ejercicio 1
Coge un texto en inglés que quieras, y calcula la frecuencia de sus palabras.



```python
# Insatalr python-docx para poder descargar archivos en word
pip install python-docx
```

```python
from docx import Document
# Cargar el documento Word
documento = Document('/Users/Esteban/OneDrive/DATOS E/Curso 2309 DATA SCIENCE (IT-ACAD Sept 23 a - 24)/Sprint 9.- Análisis del SENTIMIENTO/TASCA M9 T01/Texto OurManyTrespasses1-22rev.docx')
documento
```

```python
# Extraer el texto del documento

texto = ""# Inicializo una cadena vacía llamada Texto

for paragraph in documento.paragraphs: # Recorro todos los párrafos del documento
    texto += paragraph.text + "\n" # Agrego cada párrafo a la cadena Texto y le añado salto de línea con \n 

# Imprimir todo el texto extraído:    
texto   
```

```python
# Importar librerías que voy a necesitar: 

import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
# Tokenizar el texto en palabras con word_tokenize convirtiendo las mayúsculas en minúsculas.
# Si quisiera mantener las mayúsculas del texto original, ultilizaría el código: palabras = word_tokenize(texto)
palabras = word_tokenize(texto.lower())
palabras
```

```python
# Obtener el número total de palabras.
numero_total_palabras = len(palabras)

# Imprimir el resultado
print(f'El número total de palabras es: {numero_total_palabras}')
```

```python
# Calculo la frecuencia de las palabras con FreqDist

from nltk import FreqDist

frecuencia_palabras = FreqDist(palabras)

for palabra, frecuencia in frecuencia_palabras.items():
    print(f'{palabra}: {frecuencia}')
```

```python
# Visualización de las frecuencias en dataframe:

# Convertir FreqDist a DataFrame
df_frecuencia = pd.DataFrame(list(frecuencia_palabras.items()), columns=['Palabra', 'Frecuencia'])

# Ordenar DataFrame por frecuencia descendente
df_frecuencia = df_frecuencia.sort_values(by='Frecuencia', ascending=False)

# Mostrar el DataFrame
df_frecuencia
```

### OBTENCIÓN DE FRECUENCIAS

El dataframe anterior ofrede las frecuencias de todas las palabras, signos de puntuación y otros elementos que aparecen en el texto

### DESARROLLO DE UN INPUT PARA LOCALIZAR LA FRECUENCIA DE UNA PALABRA

Teniendo en cuenta las frecuencias anteriores, voy a desarrollar un código de un imput que me permita buscar la frecuecia de cualquier palabra o sigino que aparezca en el texto.

```python
# Para buscar la frecuencia de una palabra en concreto:

# Solicitar al usuario que ingrese la palabra
palabra_buscada = input("Ingresa una palabra: ")

# Convertir la palabra ingresada a minúsculas
palabra_buscada = palabra_buscada.lower()

# Buscar la frecuencia en el DataFrame
frecuencia_palabra = df_frecuencia.loc[df_frecuencia['Palabra'].str.lower() == palabra_buscada, 'Frecuencia'].values

if len(frecuencia_palabra) > 0:
    # Imprimir el resultado si se encontró la palabra
    print(f'La frecuencia de la palabra "{palabra_buscada}" es: {frecuencia_palabra[0]}')
else:
    # Imprimir un mensaje si la palabra no se encuentra en el DataFrame
    print(f'La palabra "{palabra_buscada}" no se encontró en el DataFrame.')
    
```

# Ejercicio 2
Saca las stopwords y realiza stemming en tu conjunto de datos.

```python
# Filtrar palabras vacías (stopwords) en dos pasos

# 1º.- Obtengo una lista de palabras vacías (stopwords) en inglés desde la biblioteca NLTK y la guardo en stop_words.
stop_words = set(stopwords.words('english'))  

# 2ª.- Filtro las palabras del texto original de la siguiente forma: 
# 2.1.- word.lower(): Convierte cada palabra a minúsculas para asegurar uniformidad en el análisis y evitar duplicaciones
# 2.2.- word.isalpha(): Verifica si la palabra contiene solo caracteres alfabéticos (letras) y no incluye ==>
# ==> caracteres especiales o números. Esto ayuda a eliminar signos de puntuación y números del análisis.
# word.lower() not in stop_words: Verifica si la palabra no está en la lista de palabras vacías ==>
# Si está en la lista de palabras vacías, se excluye de la lista final.

palabras_filtradas = [word.lower() for word in palabras if word.isalpha() and word.lower() not in stop_words]

palabras_filtradas
```

```python
# Obtener el número total de palabras filtradas.
numero_total_palabras_filtradas = len(palabras_filtradas)

# Imprimir el resultado
print(f'El número total de palabras filtradas es: {numero_total_palabras_filtradas}')
```

### NOTA SOBRE LA REDUCCIÓN DE PALABRAS CON EL STOPWORDS: 

Observamos que el total de palabras se ha reducido: 
- de 3107 palabras totales en el texto original.
- a 1630 palabras tras realizar el stopwords.
El número total de palabras se ha reducico prácticamente al 50%.

```python
# Realizar el stemming: eliminar los sufijos de palabras, como "ing" o "ed", para reducirlas a su forma básica.

from nltk.stem import PorterStemmer

# Crear un objeto PorterStemmer para inglés
stemmer = PorterStemmer()


# Aplicar stemming a las palabras filtradas en inglés
palabras_stemmed = [stemmer.stem(palabra) for palabra in palabras_filtradas]

# Imprimir las palabras originales y sus formas stem
for original, stemmed in zip(palabras_filtradas, palabras_stemmed):
    print(f'Original: {original} | Stemmed: {stemmed}')
```

```python
# Crear un DataFrame con las palabras originales y sus formas stem
df_stemmed = pd.DataFrame({
    'Original': palabras_filtradas,
    'Stemmed': palabras_stemmed
})

# Imprimir el DataFrame
df_stemmed
```

```python
# Lemmatize the tokens

from nltk.stem import WordNetLemmatizer
    
lemmatizer = WordNetLemmatizer()
palabras_lematizadas = [lemmatizer.lemmatize(palabra) for palabra in palabras_filtradas]

# Imprimir las palabras originales y sus formas lematizadas
for original, lemmatized in zip(palabras_filtradas, palabras_lematizadas):
    print(f'Original: {original} | Lemmatized: {lemmatized}')
```

```python
# Crear un DataFrame con las palabras originales y sus formas Lematizadas
df_lemmatized = pd.DataFrame({
    'Original': palabras_filtradas,
    'lemmatized': palabras_lematizadas
})

# Imprimir el DataFrame
df_lemmatized
```

### OBSERVACIONES AL STEAMING Y LEMATIZACIÓN: 

Obtenemos dos tablas, ambas de 1530 registros, que contienen las propuestas de steaming y lematización para cada palabra.

```python
df_sentimiento.describe()
```

### CONCLUSIÓN AL TEST nº1: 

Vemos que la inmensa gran mayoría de las palabras de consideran de tipo neutro.

Procedo por tanto a calcular los porcentajes que el análisis de sensibilidad arroja para palabras positivas, negativas y neutras:

```python
# Contar el número de 1 en cada columna
num_positivos = (df_sentimiento['pos'] == 1).sum()
num_negativos = (df_sentimiento['neg'] == 1).sum()
num_neutros = (df_sentimiento['neu'] == 1).sum()
num_compounds = (df_sentimiento['compound'] == 1).sum()

# Imprimir los resultados
print(f"Número de positivos: {num_positivos}")
print(f"Número de negativos: {num_negativos}")
print(f"Número de neutros: {num_neutros}")
print(f"Número de compounds: {num_compounds}")
```

```python
# Crear un DataFrame con los recuentos
df_recuentos = pd.DataFrame({
    'Sentimiento': ['Positivos', 'Negativos', 'Neutros', 'Compounds'],
    'Recuento': [num_positivos, num_negativos, num_neutros, num_compounds]
})

# Imprimir el DataFrame con los recuentos
df_recuentos
```

```python
# Calcular el total de todas las categorías
total_general = df_recuentos['Recuento'].sum()

# Calcular el porcentaje para cada categoría
df_recuentos['Porcentaje'] = (df_recuentos['Recuento'] / total_general) * 100

# Imprimir el DataFrame con los recuentos y porcentajes
df_recuentos
```

```python
# Redondear los porcentajes a un decimal
df_recuentos['Porcentaje'] = df_recuentos['Porcentaje'].round(1)

# Agregar el símbolo '%' a los porcentajes
df_recuentos['Porcentaje'] = df_recuentos['Porcentaje'].astype(str) + '%'

# Imprimir el DataFrame con los recuentos y porcentajes formateados
df_recuentos
```

### COCLUSIÓN AL ANÁLISIS DE SENTIMIENTO POR PALABRAS.

- El 93,2% de las palabras arrojan un sentimiento neutro.
- El 2,9% de las palabras arrojan un sentimiento negativo.

### TEST nº 2
### REALIZAR EL ANÁLISIS DE SENTIMIENTOS POR FRASES EN EL TEXTO SIN PREPROCESAR

A la vista del resultado anterior, me permito realizar un análisis de sentimiento en las frases del texto, no en las palabras para ver si arroja resultados diferentes al análisis por palabras.

```python
# Hay que dividir el texto en oraciones, en lugar de palabras
import nltk

# Descargar el tokenizer de oraciones
nltk.download('punkt')

# Tokenizar el texto en oraciones
oraciones = nltk.sent_tokenize(texto)

# Imprimir las oraciones
oraciones
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

# Inicializar el analizador de sentimientos
sid = SentimentIntensityAnalyzer()

# Aplicar el análisis de sentimientos a cada oración y almacenar los resultados en una lista de diccionarios
resultados_sentimiento_oraciones = [{"Oracion": oracion, **sid.polarity_scores(oracion)} for oracion in oraciones]

# Convertir la lista de resultados a DataFrame
df_sentimiento_oraciones = pd.DataFrame(resultados_sentimiento_oraciones)

# Imprimir el DataFrame con los resultados
df_sentimiento_oraciones
```

```python
# Calcular los totales de 'neg', 'neu' y 'pos' para todas las oraciones
totales_sentimiento = df_sentimiento_oraciones[['neg', 'neu', 'pos']].sum()

# Crear un nuevo DataFrame con los totales
df_totales_sentimiento = pd.DataFrame({'Totales': totales_sentimiento})

# Imprimir el nuevo DataFrame con los totales
df_totales_sentimiento
```

```python
# Calcular el total general de 'neg', 'neu' y 'pos'
total_general = totales_sentimiento.sum()

# Calcular los porcentajes respecto al total
df_totales_sentimiento['Porcentaje'] = (totales_sentimiento / total_general * 100).round(1).astype(str) + '%'

# Imprimir el nuevo DataFrame con los totales y porcentajes
df_totales_sentimiento
```

### COCLUSIÓN AL ANÁLISIS DE SENTIMIENTO POR ORACIONES.

Según nos muestra la tabla anterior, el análisis de sentimientos por frases revela que: 
- El 91,2% de las oraciones arrojan un sentimiento neutro.
- El 4,1% de las oraciones arrojan un sentimiento negativo.
- 3l 4,7% de las oraciones arrojan un sentimiento positivo.


No hay gran diferencia a los resultados obtenidos en el análisis por palabras. Cabe señalar que el análisis por oraciones asigna algo menos al conjunto de neutralidad, y valora en mayor cantidad el número de construcciones tanto positivas como negativas.
